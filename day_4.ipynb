```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from IPython.display import Image, display

# 1️⃣ Create a 2-qubit circuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0,1], [0,1])

# 2️⃣ Simulate
backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc, shots=1024).result()
counts = result.get_counts()
print("Measurement results:", counts)

# 3️⃣ Make sure Matplotlib plots inline
%matplotlib inline

# 4️⃣ Plot the histogram
fig = plot_histogram(counts, title="Bell State Results")

# 5️⃣ Force the figure to display in Colab
plt.show()            # normal display
display(fig)          # IPython fallback

# 6️⃣ OR save and show as image (ultimate fallback)
fig.savefig("hist.png")
display(Image(filename="hist.png"))
```
