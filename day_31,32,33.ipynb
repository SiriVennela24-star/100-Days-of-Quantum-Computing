```python
# Step 1: Imports
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit.library import TwoLocal
from qiskit_algorithms import VQE
from qiskit_algorithms.optimizers import COBYLA
from qiskit_aer import Aer

# Step 2: Define a simple Hamiltonian (example)
# Example: Z ⊗ Z + X ⊗ X for a 2-qubit system
qubit_op = SparsePauliOp.from_list([
    ("ZZ", 1.0),
    ("XX", 1.0)
])

# Step 3: Define ansatz with reps=3 (depth = 3)
ansatz = TwoLocal(qubit_op.num_qubits, rotation_blocks='ry', entanglement_blocks='cz', reps=3)

# Step 4: Define optimizer and backend
optimizer = COBYLA(maxiter=100)
backend = Aer.get_backend('aer_simulator')

# Step 5: Run VQE
vqe = VQE(ansatz=ansatz, optimizer=optimizer, quantum_instance=backend)
result = vqe.compute_minimum_eigenvalue(qubit_op)

# Step 6: Print result
print("Ground state energy (real part):", result.eigenvalue.real)

```

```python
#day 33 GAN prototype
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, Aer, execute
from qiskit.circuit import Parameter
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
```

```python
# Real data: 0 = 25%, 1 = 75% (simple 1-qubit distribution)
real_probs = [0.25, 0.75]
```

```python
theta = Parameter('θ')  # learnable parameter
qc = QuantumCircuit(1,1)

qc.ry(theta, 0)  # parameterized rotation
qc.measure(0,0)  # measure qubit

# Visualize the generator circuit
qc.draw('mpl')
```

```python
backend = Aer.get_backend('aer_simulator')
shots = 1024

# Initialize theta
theta_val = 0.1
learning_rate = 0.1
iterations = 30

history = []  # store generator outputs

for i in range(iterations):
    # Bind parameter
    bound_qc = qc.bind_parameters({theta: theta_val})

    # Run the circuit
    result = execute(bound_qc, backend, shots=shots).result()
    counts = result.get_counts()

    # Compute generated probability of |1>
    p1 = counts.get('1',0)/shots

    # Store history for visualization
    history.append(p1)

    # Simple gradient-like update (move towards real_prob)
    grad = real_probs[1] - p1
    theta_val += learning_rate * grad  # update theta
```

```python
plt.plot(history, marker='o')
plt.axhline(y=real_probs[1], color='r', linestyle='--', label='Real Data')
plt.title("Mini QGAN Training: Generator Output vs Real Data")
plt.xlabel("Iteration")
plt.ylabel("Probability of |1>")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Final trained generator state
# Create a new circuit without measurement for Statevector
qc_no_measure = qc.remove_final_measurements(inplace=False)

final_state = Statevector(qc_no_measure.bind_parameters({theta: theta_val}))
plot_bloch_multivector(final_state)
```
