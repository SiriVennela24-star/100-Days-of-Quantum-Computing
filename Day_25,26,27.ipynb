```python
import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Nodes: Quantum + AI components
nodes = [
    "Classical Info (θ)",
    "Alice (Quantum Encoder)",
    "Entanglement Channel",
    "Bob (Quantum Decoder)",
    "AI Analyzer"
]

# Add nodes
for n in nodes:
    G.add_node(n)

# Edges: Flow of information
edges = [
    ("Classical Info (θ)", "Alice (Quantum Encoder)"),
    ("Alice (Quantum Encoder)", "Entanglement Channel"),
    ("Entanglement Channel", "Bob (Quantum Decoder)"),
    ("Bob (Quantum Decoder)", "AI Analyzer")
]

G.add_edges_from(edges)

# Positions for nodes
pos = {
    "Classical Info (θ)": (0, 0),
    "Alice (Quantum Encoder)": (1, 1),
    "Entanglement Channel": (2, 0),
    "Bob (Quantum Decoder)": (3, 1),
    "AI Analyzer": (4, 0)
}

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color='skyblue')
# Draw edges
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=20, edge_color='gray', width=2)
# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

plt.title("Quantum Teleportation + AI Concept Diagram", fontsize=14)
plt.axis('off')
plt.show()
```

```python
#QPE
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np

# --------------------------
# QPE-like circuit
# --------------------------
qc = QuantumCircuit(3,2)
qc.h([0,1])
qc.cx(1,2)
qc.cx(0,2)
qc.h([0,1])

# --------------------------
# Simulate statevector
# --------------------------
state = Statevector.from_instruction(qc)

# --------------------------
# Compute probabilities for each basis state
# --------------------------
probs = np.abs(state.data)**2
basis_states = [format(i, '03b') for i in range(8)]  # 3-qubit states

# --------------------------
# Bar plot visualization
# --------------------------
plt.figure(figsize=(8,5))
plt.bar(basis_states, probs, color='skyblue')
plt.xlabel("Basis States")
plt.ylabel("Probability")
plt.title("Quantum State Probabilities - 3 Qubit QPE Circuit")
plt.show()

# Optional: print amplitudes and probabilities
for i, amp in enumerate(state.data):
    print(f"|{basis_states[i]}>: amplitude = {amp:.3f}, probability = {np.abs(amp)**2:.3f}")
```

```python
# --------------------------
# QPE  Visualization
# --------------------------
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# --------------------------
# 1. QPE-like circuit
# --------------------------
qc = QuantumCircuit(3,2)
qc.h([0,1])
qc.cx(1,2)
qc.cx(0,2)
qc.h([0,1])

# --------------------------
# 2. Simulate statevector
# --------------------------
state = Statevector.from_instruction(qc)

# --------------------------
# 3. Compute probabilities
# --------------------------
probs = np.abs(state.data)**2
basis_states = [format(i, '03b') for i in range(8)]  # 3-qubit states

# --------------------------
# 4. Prototype-style bar plot (3D-like)
# --------------------------
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(len(basis_states))
y = np.zeros_like(x)
z = np.zeros_like(x)
dx = 0.5 * np.ones_like(x)
dy = 0.5 * np.ones_like(x)
dz = probs

colors = plt.cm.viridis(dz / max(dz))  # gradient colors based on height

ax.bar3d(x, y, z, dx, dy, dz, color=colors, shade=True)
ax.set_xticks(x)
ax.set_xticklabels(basis_states, fontsize=12)
ax.set_xlabel("Basis States")
ax.set_ylabel("")
ax.set_zlabel("Probability")
ax.set_title("Prototype Visualization: QPE Quantum State Probabilities", fontsize=14)

# Annotate the bars with probability values
for i, p in enumerate(probs):
    ax.text(i, 0, p + 0.01, f"{p:.2f}", ha='center', fontsize=10)

plt.show()

# --------------------------
# 5. Print amplitudes and probabilities
# --------------------------
for i, amp in enumerate(state.data):
    print(f"|{basis_states[i]}>: amplitude = {amp:.3f}, probability = {np.abs(amp)**2:.3f}")
```
