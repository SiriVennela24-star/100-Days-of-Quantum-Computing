```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
theta = 1.2  # classical info
qc = QuantumCircuit(3,3)
qc.ry(theta, 0)
qc.h(1)
qc.cx(1,2)
qc.cx(0,1)
qc.h(0)
qc.measure([0,1,2],[0,1,2])
backend = Aer.get_backend('aer_simulator')
result = backend.run(qc, shots=1024).result()
print(result.get_counts())
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Nodes: Quantum + AI components
nodes = [
    "Classical Info (θ)",
    "Alice (Quantum Encoder)",
    "Entanglement Channel",
    "Bob (Quantum Decoder)",
    "AI Analyzer"
]

# Add nodes
for n in nodes:
    G.add_node(n)

# Edges: Flow of information
edges = [
    ("Classical Info (θ)", "Alice (Quantum Encoder)"),
    ("Alice (Quantum Encoder)", "Entanglement Channel"),
    ("Entanglement Channel", "Bob (Quantum Decoder)"),
    ("Bob (Quantum Decoder)", "AI Analyzer")
]

G.add_edges_from(edges)

# Positions for nodes
pos = {
    "Classical Info (θ)": (0, 0),
    "Alice (Quantum Encoder)": (1, 1),
    "Entanglement Channel": (2, 0),
    "Bob (Quantum Decoder)": (3, 1),
    "AI Analyzer": (4, 0)
}

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color='skyblue')
# Draw edges
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=20, edge_color='gray', width=2)
# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

plt.title("Quantum Teleportation + AI Concept Diagram", fontsize=14)
plt.axis('off')
plt.show()
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

# --------------------------
# 1. Circuit diagram
# --------------------------
qc.draw(output='mpl', scale=1.5)

# --------------------------
# 2. Bloch sphere visualization (before measurement)
# --------------------------
# Remove measurements temporarily to visualize quantum states
qc_no_meas = qc.remove_final_measurements(inplace=False)

# Simulate statevector
state = Statevector.from_instruction(qc_no_meas)

# Plot Bloch spheres for all qubits
plot_bloch_multivector(state)
plt.show()

# --------------------------
# 3. Print measurement counts
# --------------------------
counts = backend.run(qc, shots=1024).result().get_counts()
print("Measurement outcomes:", counts)
qc.draw('mpl')
```
